```python
import datetime
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
#variáveis globais
BASE = 'BASE.csv'
DIA_INICIO_PRED = datetime.date(2017,1,1)
DIA_FINAL_PRED = datetime.date(2023,12,31)

#configurações globais
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.2f}'.format
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv(BASE)
```

```python
#correção dos nomes das colunas
lista_nomes_corretos = []
lista_nomes_errados = df.columns

for i in range(len(lista_nomes_errados)):
    lista_nomes_corretos.append(lista_nomes_errados[i].replace(" ",""))
    
df.columns = lista_nomes_corretos
df.info()
```

```python
#transformação das strings em datetime

coluna_max = len(df.columns)
linha_max = len(df)

linha = 0
coluna = 0

for coluna in tqdm(range(coluna_max)):
    for linha in range(linha_max):
        dia_string = df.iloc[linha,coluna]
        if (dia_string is np.nan):
            pass
        else:    
            dia_obj = datetime.datetime.strptime(dia_string, '%d/%m/%Y').date()
            df.iloc[linha, coluna] = dia_obj
```

```python
def extrair_indice_individual(indice, df): 
    '''
    Extração do índice em específico da base e tratamento.
    '''
    df_fx = df.loc[:,indice]
    df_fx.dropna(inplace=True)
    df_fx = df_fx.sort_values()
    df_fx.reset_index(drop=True, inplace=True)
    
    return df_fx.to_list()
```

```python
def criar_dataframe_individual(data_inicial, data_final):
    '''
    Criar o dataframe usado para a predição da data.
    '''       
    df_fx = pd.DataFrame(pd.date_range(data_inicial, data_final), columns=['data'])
    df_fx['filtro1_dia_semana'] = 0
    df_fx['filtro2_feriado'] = 0
    df_fx['predicao'] = 0
    df_fx[indice] = 0
    df_fx['metrica'] = 0
    
    return df_fx
```

```python
def preencher_datas_conhecidas(df_fx, lista_datas_orig):
    '''
    Popular o dataframe criado com as datas conhecidas na base original.
    '''   
    for linha in range(len(df_fx)):
        if df_fx.iloc[linha,0] in lista_datas_orig:
            df_fx.iloc[linha,4] = 1
            
    return df_fx
```

```python
def preencher_dias_semana(df_fx):
    '''
    Preencher o dataframe com os dias da semana
    
    Interpretação:
    1: segunda-feira
    2: terça-feira
    3: quarta-feira
    4: quinta-feira
    5: sexta-feira
    6: sábado
    7: domingo    
    '''
    for linha in range(len(df_fx)):
        data = df_fx.iloc[linha,0]
        df_fx.iloc[linha,1] = data.isoweekday()
    
    return df_fx
```

```python
def identificar_lista_feriados(df_fx, indice, timestamp_2017):
    '''
    Identificar os feriados entre as datas da base original.
    '''
    df_feriados = df_fx[(~df_fx['filtro1_dia_semana'].isin([6,7])) &\
                        (df_fx[indice] == 0) &\
                        (df_fx['data'] < timestamp_2017)]
    
    lista_feriados = df_feriados['data'].to_list()
    
    return lista_feriados
```

```python
def ajustar_lista_feriados(lista_feriados):
    '''
    Trazer os feriados para o mesmo ano de (digamos, 2000) e eliminar os repetidos.
    '''
    lista_fx = []
    for data in lista_feriados:
        lista_fx.append(datetime.date(2000, data.month, data.day))
                        
    return set(lista_fx)
```

```python
def preencher_feriados(lista_feriados, df_fx):
    '''
    Preencher o dataframe com os feriados de cada índice.
    '''
    for linha in range(len(df_fx)):
        data = df_fx.iloc[linha,0]
        if (datetime.date(2000, data.month, data.day) in lista_feriados):
            df_fx.iloc[linha,2] = 1
    return df_fx
```

```python
def metrica(df_fx, DIA_INICIO_PRED, df_met):
    '''
    Quantifica os acertos da predição.
    '''
 
    data = df_fx.iloc[0,0]
    linha = 0
    
    while data < DIA_INICIO_PRED:
        pred = df_fx.iloc[linha,3]
        real = df_fx.iloc[linha,4]
        if (pred == real):
            df_fx.iloc[linha,5] = 1
        else:
            df_fx.iloc[linha,5] = -1
        linha += 1
        data = df_fx.iloc[linha,0]
    
    acertos = df_fx['metrica'].value_counts()[1]
    erros = df_fx['metrica'].value_counts()[-1]
    nova_linha = {'indice':indice, 'acertos':acertos, 'erros':erros}
    df_met = df_met.append(nova_linha, ignore_index=True)    
    return df_fx, df_met
```

```python
df_pred = pd.DataFrame(index=range(15000))
timestamp_2017 = pd.to_datetime(DIA_INICIO_PRED)
df_metrica = pd.DataFrame(columns=['indice','acertos','erros'])

for indice in tqdm(df.columns):
    #função de extração do índice e tratamento dos dados
    lista_datas_orig = extrair_indice_individual(indice, df)
    
    data_inicial = lista_data_orig[0]
    
    #criando o dataframe do indice
    df_indice = criar_dataframe_individual(data_inicial, DIA_FINAL_PRED)
    df_indice = preencher_datas_conhecidas(df_indice, lista_datas_orig)
    
    #incluindo dias da semana
    df_indice = preencher_dias_semana(df_indice)
    
    #identificando e incluindo feriados
    lista_feriados = identificar_lista_feriados(df_indice, indice, timestamp_2017)
    lista_feriados = ajustar_lista_feriados(lista_feriados)
    df_indice = preencher_feriados(lista_feriados, df_indice)
    
    #predicao
    df_indice.loc[(df_indice['filtro1_dia_semana'].isin(list(range(1,6)))) &\
                  (df_indice['filtro2_feriado'] == 0), 'predicao'] = 1
    
    #metrica
    df_indice, df_metrica = metrica(df_indice, DIA_INICIO_PRED, df_metrica)
    
    #organizando resultados
    df_pred_ind = df_indice[df_indice['predicao'] == 1]['data'].reset_index(drop=True)
    df_pred[indice] = df_pred_ind
```

```python
#Exportando os resultados
df_metrica.to_csv('metrica.csv', sep=';')
df_pred.to_csv('datas_pred.csv', sep=';')
```
